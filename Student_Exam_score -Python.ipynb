# PROJECT ON STUDENT EXAM SCORE 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as p
import seaborn as sns
```

```python
std=pd.read_csv("C:\\Users\\toysa\\Downloads\\Data science [ VIDEO AND PYTHON FILES]\\Student_Exam_Score.csv")
```

```python
std
```

```python
# Drop column name S.no.
std.drop(columns=['S.No. '], inplace=True)
```

```python
# Display the number of columns and rows
np.shape(std)
```

```python
# Display all of the columns 
std.columns
```

```python
# Display the details information about the dataset
std.info()
```

```python
# Count the total number of null values in the dataset
std.isnull().sum()
```

```python
# Change the WklyStudyHours columns values where 10-may
std["WklyStudyHours"]=std["WklyStudyHours"].str.replace("10-May", "5-10")
```

```python
# Summarize the data
std.describe()
```

```python
# Summary statistics for the scores
std[['MathScore', 'ReadingScore', 'WritingScore']].describe()
```

```python
# Average scores by gender
std.groupby('Gender')[['MathScore', 'ReadingScore', 'WritingScore']].mean()
```

```python
# Average scores by test preparation status
std.groupby('TestPrep')[['MathScore', 'ReadingScore', 'WritingScore']].mean()
```

```python
# Average scores by test preparation status
std.groupby('WklyStudyHours')[['MathScore', 'ReadingScore', 'WritingScore']].mean()
```

```python
# Average scores based on the number of siblings
std.groupby('NrSiblings')[['MathScore', 'ReadingScore', 'WritingScore']].mean()
```

```python
# Average scores based on the PracticeSport
std.groupby("PracticeSport")[['MathScore', 'ReadingScore', 'WritingScore']].mean()
```

```python
# Average scores based on the IsFirstChild
std.groupby("IsFirstChild")[['MathScore', 'ReadingScore', 'WritingScore']].mean()
```

```python
# Average scores based on the TransportMeans
std.groupby("TransportMeans")[['MathScore', 'ReadingScore', 'WritingScore']].mean()
```

# Visualizataion 

```python
# Display the entrie data or set or rows according to gender wise
p.figure(figsize=(7,5))
a=sns.countplot(data=std, x="Gender")
a.bar_label(a.containers[0])
p.title("DISPLAY CHART ON GENDER WISE")
p.show()
```

```python
# analysis according to Parent Education impact on study of child 
pe=std.groupby('ParentEduc')[['MathScore', 'ReadingScore', 'WritingScore']].mean()
```

```python
pe
```

```python
# Display education statue of parents on marks using heatmap
p.figure(figsize=(5,5))
sns.heatmap(pe, annot=True)
p.title("PARENT EDUCATION STATUS")
p.show()
```

```python
# Average scores by ParentMaritalStatus
pe2=std.groupby('ParentMaritalStatus')[['MathScore', 'ReadingScore', 'WritingScore']].mean()
```

```python
pe2
```

```python
# Display parent marital status on education 
p.figure(figsize=(5,5))
sns.heatmap(pe2,annot=True)
p.title("PARENT MARITAL STATUS IMAPCT ON STUDENT EDUCATION")
p.show()
```

```python
# Analysis through outlier on subject wise
sns.boxplot(data=std)
p.show()
```

```python
# apply visaulzation on Ethnic group using pie chart
# Find unique value
print(std["EthnicGroup"].unique())
```

```python
groupA=std.loc[(std["EthnicGroup"] == "group A")].count()
groupB=std.loc[(std["EthnicGroup"] == "group B")].count()
groupC=std.loc[(std["EthnicGroup"] == "group C")].count()
groupD=std.loc[(std["EthnicGroup"] == "group D")].count()
groupE=std.loc[(std["EthnicGroup"] == "group E")].count()
```

```python
print(groupA, "\n", groupB)
```

```python
# pie chart on Ethnic group of students
ethnic=["Group A", "Group B", "Group C", "Group D", "Group E"]
mlist=[groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
p.pie(mlist, labels=ethnic, autopct="%1.1f%%", explode=[0, 0, 0, 0, 0.1])
p.show()
```

```python
# Countplot to count number of student by their ethnic group
ax=sns.countplot(data=std, x="EthnicGroup")
ax.bar_label(ax.containers[0])
p.show()
```

```python
# Compare writing and reading score with gender
sns.relplot(data=std, x="WritingScore", y="ReadingScore", col="Gender")
p.show()
```

```python
# see the impact of weekly study hours on writing and reading score
sns.displot(data=std, x="WritingScore", y="ReadingScore", col="WklyStudyHours")
```

```python
# see the impact of Test Preparation on writing and reading score
sns.displot(data=std, x="WritingScore", y="ReadingScore", col="TestPrep")
```

```python
# see the impact of Practice Sport on writing and reading score
sns.displot(data=std, x="WritingScore", y="ReadingScore", col="IsFirstChild")
```
